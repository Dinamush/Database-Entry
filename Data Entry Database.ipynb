```python
import pandas as pd


def take_information():
    f_name = input("What is your first name?")
    l_name = input("What is your last name?")
    age = input("How old are you?")
    quote = input("What is your favourite quote?")
    salary = input("What is your salary?")
    balance = input("What is the account balance?")
    return {'first_name':f_name, 'last_name':l_name,'age':age,'quote':quote,'salary':salary,'balance':balance}
    
def record_information(df):
    information_dict = take_information()
    df = df.append(information_dict, ignore_index=True)
    return df
 
def update_csv(df):
    df.to_csv("database.csv", index=False)

def main():

    df = pd.read_csv('database.csv')
    df = record_information(df)  # Capture the updated DataFrame
    update_csv(df)
    
if __name__ == "__main__":
    main()
    
    
```

```python
import pandas as pd

class DatabaseManager:
    def __init__(self, dataframe):
        self.df = dataframe

    def query_account(self, name):
        name = name.capitalize()
        return self.df.query('first_name==@name')
    
    def update_account_balance_in_database(self, account_instance):
        condition = (self.df['first_name'] == account_instance.name)
        self.df.loc[condition, 'balance'] = account_instance.balance
        self.df.to_csv("database.csv", index=False)
        
    def update_account_salary(self, account_instance):
        condition = (self.df['first_name'] == account_instance.name)
        self.df.loc[condition, 'salary'] = account_instance.salary
        self.df.to_csv("database.csv", index=False)
        
class Account:
    def __init__(self, database_manager):
        #the attribute database manager is assigned to the Class DatabaseManager that opens up the database for use
        self.database_manager = database_manager
        self.retrieved = False
        self.name = None
        self.last_name = None
        self.age = None
        self.quote = None
        self.salary = None
        self.balance = None

    def enter_first_name(self, name):
        #allows user to enter their first name to check for their account
        self.name = name

    def retrieve_details(self):
        #collects all the details from the database with the given name 
        if not self.name:
            print("No first name entered.")
            return

        entry = self.database_manager.query_account(self.name)

        if entry.empty:
            print(f"No details found for {self.name}.")
            return

        self.last_name = entry['last_name'].values[0]
        self.age = entry['age'].values[0]
        self.quote = entry['quote'].values[0]
        self.salary = entry['salary'].values[0]
        self.balance = entry['balance'].values[0]
        self.retrieved = True

        
    def return_details(self):
        #this gives the details to the user 
        if not self.retrieved:
            print("Details not retrieved... Retrieving")
            self.retrieve_details()
        
        print(f'The last name is {self.last_name}, of age {self.age}, favourite quote {self.quote}, of salary ${self.salary} and account balance ${self.balance}.')

    def change_funds(self, value):
        change = "Deducted" if value < 0 else "Added"
        #negative values can be used to deduct from the account balance
        if isinstance(value, (int, float)):
            self.balance += value
            print(f"Successfully {change} ${value}")
            # Update the account balance in the database
            self.database_manager.update_account_balance_in_database(self)
            
        else:
            print("Invalid entry")

            
    def change_salary(self, new_salary):
        #changes the salary and uses the database manager class to update the account salary in the database. 
        print("Invalid Salary") if new_salary < 0 else print("New Salary Updated")
        self.salary = new_salary
        self.database_manager.update_account_salary(self)
        
```
